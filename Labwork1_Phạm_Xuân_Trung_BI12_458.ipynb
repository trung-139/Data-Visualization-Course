```python
import pandas as pd

# Read the data from the text file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Process the lines to create a list of lists
data = []
for line in lines:
    data.append(list(map(float, line.strip().split())))

# Create a DataFrame
df = pd.DataFrame(data)

# Delete the first row and first column
df = df.iloc[1:, 1:]


df = df.transpose()
df.to_csv("new_data.csv")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a pandas DataFrame
data = pd.read_csv('new_data.csv', index_col=0)

# Replace -99.0 with NaN for better visualization
data = data.replace(-99.0, np.nan)

# Calculate cumulative days in each month
days_in_month = [0] + [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
cumulative_days = np.cumsum(days_in_month)

# Set the figure size
plt.figure(figsize=(10, 3))

# Plot the data
plt.plot(data.values.flatten(), label='Temperature', linestyle='-')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature from 01 Jan 2008 to 31 Dec 2008')

# Set y-axis tick intervals
plt.yticks(np.arange(5, 40, 5))

# Set x-axis tick positions and labels
plt.xticks(cumulative_days[:-1], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Display grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Save the figure
plt.savefig('/content/drive/MyDrive/Figure1_1.png')

# Show the plot
plt.show()
```
